```python
import pandas as pd
import numpy as np
```

a. Extraer información del archivo

b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean de tipo *datetime64*

```python
vaccs = pd.read_csv('/content/drive/MyDrive/Data Science (Arnoldo Ramirez) - Learning Gate/02 Pandas & Numpy/country_vaccinations.csv', parse_dates=['date'])

vaccs.info()
```

-------------------------------
c. Determinar la cantidad de vacunas aplicadas de cada compañía (Con base en cómo lo reporta cada país en la columna vaccines)

```python
#Nota: Para resolver este ejercicio fue necesario leer un archivo distinto
#al que se indica en las instrucciones
vaccs_by_manufacturer = pd.read_csv('/content/drive/MyDrive/Data Science (Arnoldo Ramirez) - Learning Gate/02 Pandas & Numpy/country_vaccinations_by_manufacturer.csv')

vaccs_by_manufacturer[['vaccine', 'total_vaccinations']].groupby(vaccs_by_manufacturer['vaccine']).sum()
```

-------------------------
d. Obtener la cantidad de vacunas aplicadas en todo el mundo

```python
totalvaccs = vaccs['daily_vaccinations_raw'].sum()

print('El total de vacunas aplicadas en todo el mundo es ', totalvaccs)
```

-----------------------------------
e. Calcular el promedio de vacunas aplicadas por país

```python
vaccs[['country', 'total_vaccinations']].groupby(vaccs['country']).mean()
```

----------------------------
f. Determinar la cantidad de vacinas aplicadas el día 29/01/21 en todo el mundo

```python
vaccs290121 = vaccs[vaccs['date'] == pd.Timestamp("2021-01-29")]
print('El total de vacunas aplicadas el día 29/01/2021 en todo el mundo fue de:', vaccs290121['daily_vaccinations_raw'].sum() )
```

------------------
g. Crear un dataframe nuevo denominado **conDiferencias** que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccinations y daily_vaccinations_raw

```python
conDiferencias = vaccs
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']

conDiferencias
```

---------------------------
h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua

```python
print('Registro más reciente:', conDiferencias['date'].max())
print('Registro más antiguo:', conDiferencias['date'].min())
```

----------------------------
i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la volumna vaccines y separar por el carácter /)

```python
conCantidad = vaccs

conCantidad['canVac'] = conCantidad['vaccines'].str.split(', ').str.len()

conCantidad[['country', 'canVac']].groupby(conCantidad['country']).mean()
```

-------------
j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020. 

```python
antes20 = vaccs[vaccs['date'] < pd.Timestamp("2020-12-20")]

antes20.info()
```

--------------
k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer. 

```python
pfizer = vaccs[vaccs['vaccines'].str.contains('Pfizer', case=False)]

pfizer.info()
```

-----------------
l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter. 

```python
writer = pd.ExcelWriter('/content/drive/MyDrive/Data Science (Arnoldo Ramirez) - Learning Gate/02 Pandas & Numpy/resultadoReto.xlsx')

conDiferencias.to_excel(writer, sheet_name='conDiferencias')
conCantidad.to_excel(writer, sheet_name='conCantidad')
antes20.to_excel(writer, sheet_name='antes20')
pfizer.to_excel(writer, sheet_name='pfizer')

writer.save()
```
